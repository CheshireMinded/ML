```python
import os
import pandas as pd

#path to dataset dir
dataset_path = "../input/video-game-sales-analyze"

#file name
file_name = "vgsales new.csv"

#full path
file_path = os.path.join(dataset_path,file_name)

#load dataset
data = pd.read_csv(file_path)

#preview the dataset
print("Dataset Preview:")
print(data.head())
```

```python
# View rows and columns
rows, cols = data.shape
print(f"The dataset has {rows} rows and {cols} columns.")
```

```python
# Missing values in each column
missing_values = data.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# Unique genres
unique_genres = data['Genre'].unique()
print("Unique genres in the dataset:")
print(unique_genres)
```

```python
# Publisher with the highest global sales
publisher_sales = data.groupby('Publisher')['Global_Sales'].sum()
top_publisher = publisher_sales.idxmax()
top_publisher_sales = publisher_sales.max()

print(f"The publisher with the highest global sales is {top_publisher} with {top_publisher_sales:.2f} million sales.")
```

```python
# Game with the highest global sales
highest_sales_game = data.loc[data['Global_Sales'].idxmax()]
print("The game with the highest global sales is:")
print(highest_sales_game)
```

```python
# Get the top game in global sales for each Platform and Genre
top_game = (
    data.loc[data.groupby(['Platform', 'Genre'])['Global_Sales'].idxmax()]
    .reset_index(drop=True)
)

# Display the result
print("Top game in global sales for each Platform and Genre:")
print(top_game[['Platform', 'Genre', 'Name', 'Global_Sales']])
```
